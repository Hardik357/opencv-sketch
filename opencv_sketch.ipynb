```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
image = cv.imread('/content/hardik.png')
cv2_imshow(image)
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
invrt_img = cv.bitwise_not(gray)
cv2_imshow(invrt_img)
```

```python
gblur_img=cv.GaussianBlur(invrt_img,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(gblur_img)
```

```python
def dodge_img(x,y):
    return cv.divide(x,255-y,scale=256)
dodged_img=dodge_img(gray,gblur_img)
cv2_imshow(dodged_img)
```

```python
def burn_img(image, mask):
    return 255 - cv.divide(255-image, 255-mask, scale=256)
final_image=burn_img(dodged_img,gblur_img)
cv2_imshow(final_image)
```
